```python
#라이브러리 로드
import numpy as np
import pandas as pd
from tqdm import tqdm
import cv2
import os
import random
import matplotlib.pyplot as plt
import shutil

from keras.preprocessing.image import load_img 

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

## 0. 상숫값 및 경로 설정

```python
data_dir = 'saves/' #1.데이터 불러오기 - csv 데이터 폴더 디렉토리
image_dir = "images/" #5.이미지 로드 및 그룹화 - 이미지 폴더 디렉토리
result_dir = "result1/" #7. 데이터 저장 - 데이터 저장 디렉토리
pca_threshold = 0.99 #3. PCA - 적합한 차원 수 찾기
```

## 1. 데이터 불러오기

```python
# this list holds all the image filename
file_names = []
data = []

with os.scandir(data_dir) as files:
    for file in files:
        # 모든 csv 파일 읽어오기
        if file.name.endswith('csv'):
            file_names.append(data_dir + file.name)
            
for file in file_names:
    now = pd.read_csv(file, header = 0, index_col = 0).to_numpy()
    data.append(now)
```

## 2. RGB 값의 벡터화

```python
def flatten_data_with_rgb(data):
    div = 0.20001
    flt_data = []
    for feature in data:
        r = (feature[:,0] / div).astype(int)
        g = (feature[:,1] / div).astype(int)
        b = (feature[:,2] / div).astype(int)
        #Green > Red > Blue 
        tar = g * 25 + r * 5 + b
        
        new_row = np.zeros(125)
        for num, val in zip(tar, data[0][:, 6]):
            new_row[num] += val + (0.01 * num)
            
        flt_data.append(new_row)
    return flt_data
```

```python
def flatten_data_with_hsv(data):
    div = 0.20001
    flt_data = []
    for feature in data:
        r = (feature[:,3] / div).astype(int)
        g = (feature[:,4] / div).astype(int)
        b = (feature[:,5] / div).astype(int)
        tar = r * 25 + g * 5 + b
        
        new_row = np.zeros(125)
        for num, val in zip(tar, data[0][:, 6]):
            new_row[num] += val
            
        flt_data.append(new_row)
    return flt_data
```

```python
flt_data = flatten_data_with_rgb(data)
```

## 3. PCA - 적합한 차원 수 찾기

```python
def pca_test(data, threshold):
    pca = PCA(n_components = len(data[0]), random_state = 2021)
    pca.fit_transform(data)
    
    ratio = np.cumsum(pca.explained_variance_ratio_)
    d = np.argmax(ratio >= threshold) + 1    
    return (d, ratio[d - 1])
```

```python
pca_component = pca_test(flt_data, pca_threshold)
pca_component
```

## 4. 클러스터링 수행

```python
def clustering_data(data, pca_components = 50, target_cluster = 16):
    pca = PCA(n_components = pca_components, random_state=2021)
    x = pca.fit_transform(data)
    
    kmeans = KMeans(n_clusters=target_cluster, random_state=22)
    kmeans.fit(x)
```

```python
clu_data = clustering_data(flt_data, pca_component[0])
```

## 5. 이미지 로드 및 그룹화

```python
image_file_names = []

for file in file_names:
    image_file_names.append(image_dir + file.rsplit('.')[0].rsplit('/')[1] + '.jpeg')
```

```python
def grouping_by_clusters(image_file_names, labels):
    groups = {}
    for file, cluster in zip(image_file_names, labels):
        if cluster not in groups.keys():
            groups[cluster] = []
            groups[cluster].append(file)
        else:
            groups[cluster].append(file)
    return groups
```

```python
groups = grouping_by_clusters(image_file_names, kmeans.labels_)
```

## 6. 데이터 시각화

```python
def view_cluster(groups, cluster):
    plt.figure(figsize = (25,25));
    # gets the list of filenames for a cluster
    files = groups[cluster]
    # only allow up to 30 images to be shown at a time
    if len(files) > 24:
        print(f"Clipping cluster size from {len(files)} to 24")
        files = files[:24]
    # plot each image in the cluster
    for index, file in enumerate(files):
        plt.subplot(8, 8,index+1);
        img = load_img(file)
        img = np.array(img)
        plt.imshow(img)
        plt.axis('off')
```

```python
view_cluster(groups, 8)
```

## 7. 데이터 저장

```python
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
```

```python
def save_data(groups, result_dir):
    for clusters, files in groups.items():
        createFolder(result_dir + str(clusters))
        for file in files:
            shutil.copyfile(file, result_dir + str(clusters) + '/' + file.rsplit('/')[1])
```

```python
save_data(groups, result_dir)
```
